```python
weights = []
with open('mwis.txt') as f:
    n = int(f.readline())
    for x in f:
        weights.append(int(x))
print(weights[0])
```

```python
#https://wincent.com/wiki/Computing_the_Maximum_Weighted_Independent_Set_of_a_graph_path

A = [0]*n
A[0] = 0
A[1] = weights[0]
for i in range(2,n):
    A[i] = max(A[i-1], A[i-2] + weights[i-1])
```

```python
mwis=set()
n=n
while n >= 1:
    if A[n - 1] >= (A[n - 2] + weights[n-1]):
        n = n - 1
    else:
        mwis.add(n)
        n = n - 2
print(mwis)
out = []
for x in [1,2,3,4,17,117,517,997]:
    if x in mwis:
        print(1)
    else:
        print(0)
```
