```python
def minkey(key,mstSet):
    m = 50000000
    index = 0
    for x in range(0,500):
        if key[x] < m and mstSet[x] ==False:
            m = key[x]
            index = x
    return index
```

```python
def prims():
    key=[50000000]*500
    parent = [None]*500
    key[0] = 0
    mstSet = [False]*500
    
    parent[0] = -1000000

    for i in range(0,500):
        min_key = minkey(key,mstSet)
        mstSet[min_key] = True
        
        for j in range(0,500):
            #print(graph[min_key][j],mstSet[j],min_key)
            if graph[min_key][j] > -1000000000000 and mstSet[j] == False and key[j] > graph[min_key][j]:
                #print(j)
                key[j] = graph[min_key][j]
                parent[j] = min_key
    return parent
```
