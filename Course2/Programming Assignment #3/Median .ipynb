```python
median =[]
entry = []
i=0
with open('Median.txt') as f:
    for x in f:
        a = int(x)
        entry.append(a)
        entry.sort()
        print(entry)
        if i%2 == 0:
            median.append(entry[int(i/2)])
        else:
            median.append(entry[int(((i+1)/2)-1)])
        print(median)
        i +=1
            
```
