```python
graph =[[0 for x in range(202)] for y in range(202)]
with open('dijkstraData.txt') as f:
    for x in f:
        entry = [y for y in x.split()]
        v1 = int(entry[0])
        for t in entry[1:]:
            v2,w = [int(u) for u in t.split(",")]
            graph[v1][v2] = w
print(graph)
```

```python
dist = [1000000]*202
dist[1] = 0
shortset = [False]*202
def mindist():
    m = 10000000
    for v in range(1,202):
        print(v)
        if dist[v] < m and shortset[v] == False:
            m = dist[v]
            index = v
    return index

for x in range(1,202):
    #print(x)
    u = mindist()
    shortset[u] = True
    for v in range(1,202):
        if graph[u][v] > 0 and shortset[v] == False and dist[v] > dist[u]+graph[u][v]:
            dist[v] = dist[u] + graph[u][v]
```
